```python
from pyAudioAnalysis import audioBasicIO
from pyAudioAnalysis import audioFeatureExtraction
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

%matplotlib inline
plt.rcParams['figure.dpi']= 90
```

```python
def getFeatures(file):
    [Fs, x] = audioBasicIO.readAudioFile(file);
    
    nSamples = int(Fs * 29.9)
    x = x[:nSamples]
    
    # F, f_names = audioFeatureExtraction.stFeatureExtraction(x, Fs, 0.050*Fs, 0.025*Fs);
    mterm, sterm, f_names = audioFeatureExtraction.mtFeatureExtraction(x, Fs, 1*Fs, 1*Fs, 0.050*Fs, 0.025*Fs);
    return mterm, f_names

mterm1, f_names1 = getFeatures("./music_positive/output_Tempest 3.wav")
mterm2, f_names2 = getFeatures("./music_negative/000002.wav")
mterm3, f_names3 = getFeatures("./music_negative/010250.wav")
```

```python
# Normalizes the dataset so that each feature has mean 0 and deviation 1
def zNormalize(featVec):
    for i in range(len(featVec)):
        featVec[i] = (featVec[i] - np.mean(featVec[i])) / np.std(featVec[i])

mterm1
```

```python
featsIdx = [0, 1, 8, 9, 10, 11]

for i in range(len(featsIdx)):
    featIdx = featsIdx[i]
    plt.figure(figsize=(12,5))
    
    plt.subplot(1,2,1)
    plt.plot(mterm1[featIdx,:], "bo-", label="Classical (fast tempo)");
    plt.plot(mterm2[featIdx,:], "r--", label="Hip Hop");
    plt.plot(mterm3[featIdx,:], "gx:", label="Instrumental");
    plt.xlabel("Window no")
    plt.ylabel(f_names1[featIdx])
    plt.legend()
    
    plt.subplot(1,2,2)
    plt.plot(mterm1[featIdx+34,:], "bo-", label="Classical (fast tempo)");
    plt.plot(mterm2[featIdx+34,:], "r--", label="Hip Hop");
    plt.plot(mterm3[featIdx+34,:], "gx:", label="Instrumental");
    plt.xlabel("Window no")
    plt.ylabel(f_names1[featIdx+34])
    plt.legend()
    
    plt.savefig(f_names1[featIdx+34] + ".png", dpi=150)
    plt.show()
```
